```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import argparse
from sqlalchemy import create_engine

class AutomatedEDA:
    def __init__(self, data_path):
        self.data_path = data_path
        self.data = None

    def load_data(self):
        if self.data_path.endswith('.csv'):
            self.data = pd.read_csv(self.data_path)
        elif self.data_path.endswith('.xlsx'):
            self.data = pd.read_excel(self.data_path)
        elif self.data_path.startswith('sqlite://'):
            engine = create_engine(self.data_path)
            self.data = pd.read_sql_query('SELECT * FROM your_table', engine)
        else:
            raise ValueError("Unsupported data format")

    def preprocess_data(self):
        self.data.dropna(inplace=True)
        
        categorical_columns = self.data.select_dtypes(include=['object']).columns
        self.data = pd.get_dummies(self.data, columns=categorical_columns)
        
        numerical_columns = self.data.select_dtypes(include=['int64', 'float64']).columns
        self.data[numerical_columns] = (self.data[numerical_columns] - self.data[numerical_columns].min()) / (self.data[numerical_columns].max() - self.data[numerical_columns].min())
        
    def generate_visualizations(self):
        for column in self.data.columns:
            if self.data[column].dtype == 'object':
                plt.figure(figsize=(10, 6))
                sns.countplot(x=column, data=self.data)
                plt.title(f'Countplot of {column}')
                plt.xticks(rotation=45)
                plt.show()
            elif self.data[column].dtype in ['int64', 'float64']:
                plt.figure(figsize=(10, 6))
                sns.histplot(data=self.data, x=column, bins=20, kde=True)
                plt.title(f'Histogram of {column}')
                plt.show()
                plt.figure(figsize=(10, 6))
                sns.boxplot(x=self.data[column])
                plt.title(f'Boxplot of {column}')
                plt.show()
                plt.figure(figsize=(10, 6))
                sns.scatterplot(x=self.data.index, y=self.data[column])
                plt.title(f'Scatterplot of {column}')
                plt.show()

    def run(self):
        self.load_data()
        self.preprocess_data()
        self.generate_visualizations()

data_path = 'C:\\Users\\hp\\Downloads\\archive (2)\\International_Report_Departures.csv'

eda_tool = AutomatedEDA(data_path)
eda_tool.run()
```
